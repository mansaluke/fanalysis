```python
from fanalysis.Models.optimize import optimize_rf, do_rf
```

```python
def test_split_df(df, prop = 0.25):
    df.sort_values(by=['Date'], inplace=True)
    return df[:round(len(df) *(1-prop))], df[round(len(df) * (1-prop)):]

df, test = test_split_df(df)
```

```python
orf =  optimize_rf(df, n_estimators=100, indep_col = 'EURUSD.bid' )        
opt_n_estimators = orf.opt_n_estimators
rf = r.do_rf(df, n_estimators=opt_n_estimators, indep_col = 'EURUSD.bid')
rf.predict_out(True)
rf.return_error_details()
rf.print_score()
rf.importances()
```

```python
oosp = rf.out_of_sample_pred(test, False)

df_test = pd.merge(oosp, test, on='Date')    
df_test.iloc[:, 5:].head()
```

```python
df_test['mean'] =  df['EURUSD.bid'].mean()

print(abs((df_test['EURUSD.bid']-df_test['mean']).mean()) - abs((df_test['EURUSD.bid']-df_test['oos_pred']).mean()))
```
